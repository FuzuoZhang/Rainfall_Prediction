```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('hourly-weather-surface/station_385_small.csv')
data = data.drop(columns=['Unnamed: 0'])
data
```

```python
N = data.shape[0]

n_train = round(N*0.7)
n_valid = round(N*0.2)
n_test = N-n_train-n_valid
```

```python
train = data.iloc[:n_train]
valid = data.iloc[n_train:n_train+n_valid]
test = data.iloc[n_train+n_valid:]
```

### 对每个数据集按照滑动窗的方法前T个时刻的所有特征作为一个sample，target为下一时刻的降雨量

```python
T = 6
```

```python
def datatrans(data, T):
    n,p = data.shape
    n_sample = n-T
    newdata = np.zeros((n_sample,T,p))
    target = np.zeros((n_sample,))
    for i in range(n_sample):
        newdata[i] = data.iloc[i:i+T]
        target[i] = data.iloc[i+T,12]
    return newdata, target
    
```

```python
X_train, y_train = datatrans(train, T)
X_valid, y_valid = datatrans(valid, T)
X_test, y_test = datatrans(test, T)
```

```python
print(X_train.shape,y_train.shape)
print(X_valid.shape, y_valid.shape)
print(X_test.shape, y_test.shape)
```

```python
print(np.mean(y_train))
print(np.mean(y_valid))
print(np.mean(y_test))
```

### Xgboost

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

将数据按照时间维度展开

```python
t,p = X_train.shape[1:]
X_train = np.reshape(X_train,(X_train.shape[0], t*p))
X_valid = np.reshape(X_valid,(X_valid.shape[0], t*p))
X_test = np.reshape(X_test,(X_test.shape[0], t*p))
print(X_train.shape)
print(X_valid.shape)
print(X_test.shape)
```

```python
model = xgb.XGBRegressor(max_depth=5, 
                         learning_rate=0.01,
                         n_estimators=500,
                         booster='gbtree', 
                         nthread =-1,
                         gamma = 0.1,
                         subsample = 0.7,
                         colsample_bytree = 0.7,
                         colsample_bylevel = 1,
                         silent=False, 
                         reg_alpha = 2,
                         reg_lambda = 0,
                         min_child_weight=2
                         )
eval_set = [(X_valid, y_valid)]
model.fit(X_train, y_train, eval_set=eval_set, verbose=True)
```

```python
y_valid_xgb = model.predict(X_valid)
mse = mean_squared_error(y_valid,y_valid_xgb)
mae = mean_absolute_error(y_valid, y_valid_xgb)
r2 = r2_score(y_valid, y_valid_xgb)
print('/********验证集实验**********/')
print('mean squared error is: {}'.format(mse))
print('mean absolute error is: {}'.format(mae))
print('R2 score is: {}'.format(r2))
```

```python
def model_predict(model, X, y):
    y_pre = model.predict(X)
    mse = mean_squared_error(y, y_pre)
    mae = mean_absolute_error(y, y_pre)
    r2 = r2_score(y, y_pre)
    print('/********预测实验**********/')
    print('mean squared error is: {}'.format(mse))
    print('mean absolute error is: {}'.format(mae))
    print('R2 score is: {}'.format(r2))
```

```python
model_predict(model, X_train, y_train)
```

将原有的y值分段离散化改为分类标签

```python
def value2class(y):
    y[np.where((y>0)&(y<=1))]=1
    y[np.where((y>1)&(y<=4))]=2
    y[np.where((y>4)&(y<=16))]=3
    y[np.where(y>16)]=4
    return y.astype('int')

y_train = value2class(y_train)
y_valid = value2class(y_valid)
y_test = value2class(y_test)
```
