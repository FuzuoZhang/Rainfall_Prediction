```python
N = data.shape[0]

n_train = round(N*0.7)
n_valid = round(N*0.2)
n_test = N-n_train-n_valid
```

```python
train = data.iloc[:n_train]
valid = data.iloc[n_train:n_train+n_valid]
test = data.iloc[n_train+n_valid:]
```

### 对每个数据集按照滑动窗的方法前T个时刻的所有特征作为一个sample，target为下一时刻的降雨量

```python
T = 6
```

```python
def datatrans(data, T):
    n,p = data.shape
    n_sample = n-T
    newdata = np.zeros((n_sample,T,p))
    target = np.zeros((n_sample,))
    for i in range(n_sample):
        newdata[i] = data.iloc[i:i+T]
        target[i] = data.iloc[i+T,12]
    return newdata, target
    
```

```python
X_train, y_train = datatrans(train, T)
X_valid, y_valid = datatrans(valid, T)
X_test, y_test = datatrans(test, T)
```

```python
print(X_train.shape,y_train.shape)
print(X_valid.shape, y_valid.shape)
print(X_test.shape, y_test.shape)
```

### Xgboost

将数据按照时间维度展开

```python
t,p = X_train.shape[1:]
X_train = np.reshape(X_train,(X_train.shape[0], t*p))
X_valid = np.reshape(X_valid,(X_valid.shape[0], t*p))
X_test = np.reshape(X_test,(X_test.shape[0], t*p))
print(X_train.shape)
print(X_valid.shape)
print(X_test.shape)
```

```python
model = xgb.XGBRegressor(max_depth=10, learning_rate=0.1, n_estimators=160, silent=False, objective='reg:gamma')
model.fit(X_train, y_train)
```

```python
y_test_xgb = model.predict(X_test)
mes = mean_squared_error(y_test_xgb, y_test)
mes
```
